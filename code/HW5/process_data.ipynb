```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Load an audio file
audio_path = r'C:\Computer Science Programs\Fall_2024\EE502_BioMed\project\data\cleaned\19\198\19-198-0000.wav'  # Replace with your file path
y, sr = librosa.load(audio_path, sr=None)  # sr=None preserves the original sample rate

# Compute MFCCs
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)  # Extract 13 MFCCs

# Plot the MFCCs
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfccs, x_axis='time', sr=sr)
plt.colorbar(format='%+2.0f dB')
plt.title('MFCC')
plt.tight_layout()
plt.show()
```
