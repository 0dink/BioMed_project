```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
audio_path1 = r'C:\Computer Science Programs\Fall_2024\EE502_BioMed\project\data\extracted_words\that_no_extra_time\87\87-121553-0003_1.wav'  # Replace with the path to your audio file
audio_path2 = r'C:\Computer Science Programs\Fall_2024\EE502_BioMed\project\data\extracted_words\that_no_extra_time\87\87-121553-0000_1.wav'
audio_path3 = r'C:\Computer Science Programs\Fall_2024\EE502_BioMed\project\data\extracted_words\that_no_extra_time\87\87-121553-0004_1.wav'

y1, sr1 = librosa.load(audio_path1, sr=None)
y2, sr2 = librosa.load(audio_path2, sr=None)
y3, sr3 = librosa.load(audio_path3, sr=None)



if sr1 != sr2:
    raise ValueError("Sampling rates of the audio files do not match.")

# Plot the waveforms
plt.figure(figsize=(12, 4))
librosa.display.waveshow(y1, sr=sr1, color='red', alpha=0.7, label='Audio 1')
librosa.display.waveshow(y2, sr=sr2, color='blue', alpha=0.7, label='Audio 2')
librosa.display.waveshow(y3, sr=sr3, color='green', alpha=0.7, label='Audio 2')
plt.title('Overlayed Waveforms')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```
